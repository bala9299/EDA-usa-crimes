### Dates and Time

* Around 1 AM maximum number of the crimes were done.
* Around 5 AM minimum number of crimes were done. 
* The most number of crimes were done after 12 PM.
* Almost every day the crimes were recorded equally but the crime rates are little low in weekend compared to weekdays.
* The most number of crimes were done in JULY month.
* The most number of crimes were done in 2017 and 2018.





```python
df.Start_Date_Time
```

```python
df.Start_Date_Time = pd.to_datetime(df.Start_Date_Time)
df.Start_Date_Time
```

```python
sns.set_style('darkgrid')
sns.distplot(df.Start_Date_Time.dt.hour,bins=24 , kde=False, norm_hist=True )
plt.title('crimes by hours');
```

```python
sns.distplot(df.Start_Date_Time.dt.dayofweek,bins=7 , kde=False , norm_hist=True)
plt.title('crimes by weekdays');
```

```python
sns.distplot(df.Start_Date_Time.dt.month,bins=12 , kde=False , norm_hist=True)
plt.title('crimes by month');
```

```python
sns.distplot(df.Start_Date_Time.dt.year,bins=7 , kde=False , norm_hist=True)
plt.title('crimes by year');
```

###### Reason why crimes were low in 2016 and 2022
There are missing datas in 2016 and 2022.
